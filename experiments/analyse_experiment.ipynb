```python
import pandas as pd

df = pd.read_csv('../experiments/results.csv')
```

```python
df[df['Smoothing_dimension'] == 1]
```

```python
df['Scaling'] = df['Scaling'].astype('category')
df['Sample_weights'] = df['Sample_weights'].astype('category')
df['Smoothing_algorithm'] = df['Smoothing_algorithm'].astype('category')
df['Background'] = df['Background'].astype('category')
```

```python
df['Smoothing_algorithm'].unique()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

formula = 'Mean_between_sensitivity_and_specificity ~ C(Scaling) * C(Sample_weights) * Smoothing_dimension * C(Smoothing_algorithm) * Smoothing_value * C(Background) * Background_blood * Background_light'

model = ols(formula, data=df).fit()

anova_results = sm.stats.anova_lm(model, typ=2)
anova_df = pd.DataFrame(anova_results)
pd.set_option('display.max_rows', None)
#print(anova_df)
```

```python
pd.set_option('display.float_format', '{:.6f}'.format)
#anova_df
anova_df[anova_df["PR(>F)"] < 0.05]
```

```python
mapping_scaling = {'svn': 1, 'normalization': 2}
mapping_smoothing = {'m': 1, 'g': 2, 's': 3, 'no': 4}

df['Scaling_encoded'] = df['Scaling'].map(mapping_scaling)
df['Sample_weights_encoded'] = df['Sample_weights'].astype(int)  
df['Smoothing_algorithm_encoded'] = df['Smoothing_algorithm'].map(mapping_smoothing)
df['Background_encoded'] = df['Background'].astype(int)  

y = df['Mean_between_sensitivity_and_specificity']
X = sm.add_constant(df[['Size',
                        'Scaling_encoded', 
                        'Sample_weights_encoded', 
                        'Smoothing_dimension', 
                        'Smoothing_algorithm_encoded', 
                        'Smoothing_value',
                        'Background_encoded', 
                        'Background_blood', 
                        'Background_light']])

model = sm.OLS(y, X).fit()
print(model.summary())
```
