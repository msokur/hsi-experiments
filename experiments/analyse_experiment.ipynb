<h1>ANOVA</h1>

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.formula.api import glm

#df['Smoothing_algorithm'] = df['Smoothing_algorithm'].cat.reorder_categories(['no', 's', 'm', 'g'])


#formula = 'Mean_between_sensitivity_and_specificity ~ C(Size) * C(Scaling) * C(Sample_weights) * Smoothing_dimension * C(Smoothing_algorithm) *  Smoothing_value * C(Background) * Background_blood * Background_light'
formula = 'Mean_between_sensitivity_and_specificity ~ Size + Sample_weights_encoded + Smoothing_dimension + Smoothing_algorithm_encoded +  Smoothing_value + Background_blood + Background_light'
#formula = 'Mean_between_sensitivity_and_specificity ~ Smoothing_value + C(Smoothing_value)'

model = ols(formula, data=df).fit()
model_glm = glm(formula, data=df, family=sm.families.Gaussian(sm.families.links.log())).fit()


anova_results = sm.stats.anova_lm(model, typ=2)

print(anova_results[anova_results["PR(>F)"] < 0.1])

print(model_glm.summary())
#anova_df = pd.DataFrame(anova_results)
#pd.set_option('display.max_rows', None)
#print(anova_df)


#mod = smf.ols(formula='Lottery ~ Literacy + Wealth + Region', data=df)
#res = mod.fit()
#print(res.summary())
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Q-Q Plot
residuals = model_glm.resid
fig = sm.qqplot(residuals, line='s')
plt.show()

# Shapiro-Wilk Test
shapiro_test = stats.shapiro(residuals)
print('Shapiro-Wilk Test:', shapiro_test)
```

```python
import numpy as np

from statsmodels.stats.outliers_influence import variance_inflation_factor
mapping_scaling = {'svn': 1, 'normalization': 2}
mapping_smoothing = {'m': 1, 'g': 2, 's': 3, 'no': 4}

df['Scaling_encoded'] = df['Scaling'].map(mapping_scaling)
df['Sample_weights_encoded'] = df['Sample_weights'].astype(int)  
df['Smoothing_algorithm_encoded'] = df['Smoothing_algorithm'].map(mapping_smoothing)
df['Background_encoded'] = df['Background'].astype(int)  

# Assuming df_dummies is your DataFrame with dummy variables
columns = ['Size', 'Sample_weights_encoded', 
                'Smoothing_dimension', 'Smoothing_algorithm_encoded', 'Smoothing_value', 'Background_blood', 'Background_light']
X = df[['Size', 
                'Sample_weights_encoded', 
                'Smoothing_dimension', 'Smoothing_algorithm_encoded', 'Smoothing_value', 'Background_blood', 'Background_light']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['Feature'] = X.columns
print(vif)

corr = X.corr()

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(columns)
ax.set_yticklabels(columns)
plt.show()
```
