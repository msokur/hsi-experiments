<h1>ANOVA</h1>

```python
# Clear ANOVE for the paper

import statsmodels.api as sm
from statsmodels.formula.api import ols, glm
import statsmodels.api as sm
from statsmodels.formula.api import glm
from IPython.display import display
from statsmodels.stats.outliers_influence import variance_inflation_factor
pd.set_option('display.float_format', lambda x: '%.3f' % x)

folder = 'statistical_analysis\\'

df_a = df[df['Scaling'] == 'svn']
df_a = df_a[df_a['Size'] != 7]

def print_vifs(model):
    X = model.model.exog
    vif = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
    print("VIF:", vif)

def calculate_anova(formula, name):
    model = ols(formula, data=df_a).fit()
    anova_results = sm.stats.anova_lm(model, typ=2)

    columns = ["sum_sq", "F", "PR(>F)"]
    anova_results[columns].sort_values(by="F", ascending=False).to_excel(f"{folder}{name}.xlsx")

    display(anova_results)
    print(model.summary())
    
    print_vifs(model)

def save_significant_sorted_values(df_results, condition, column, name):
    sorted_results = df_results[(df_results['P>|t|'] < 0.1) & (condition)].sort_values(by=column, ascending=True)
    display(sorted_results)
    sorted_results.to_excel(f"{folder}{name}_sorted.xlsx")
    
def calculate_ols(formula, name, sort=0):    
    # sort = 1 - show and save combinations that influance positively
    # sort = 2 - show and save comninations that influance negatively
    # sort = 0 - no additional actions
    model = ols(formula, data=df_a).fit()
    
    print(model.summary())
    
    print_vifs(model)

    results = model.summary2().tables[1]  # This fetches the regression results table
    df_results = pd.DataFrame(results)
    df_results.to_excel(f"{folder}{name}_all.xlsx")

    # for the worst results we sort be Coef., because it's more intersting mean worsening
    # for the best results we sort nach '0.975]', because we are interested in the maximal achivements of this method
    if sort == 1:
        save_significant_sorted_values(df_results, df_results['0.975]'] > 0, "0.975]", name+"_positive")
    if sort == 2:
        save_significant_sorted_values(df_results, df_results['0.975]'] <= 0, "Coef.", name+"_negative")

formula_whole = 'Mean_between_sensitivity_and_specificity ~ Size + Sample_weights_encoded + Smoothing_dimension + Smoothing_algorithm_encoded + Smoothing_value + Background_blood + Background_light'
formula_smoothing_separate = 'Mean_between_sensitivity_and_specificity ~ C(Smoothing_encoded)'
formula_blood_separate = 'Mean_between_sensitivity_and_specificity ~ C(Background_blood)'
formula_light_separate = 'Mean_between_sensitivity_and_specificity ~ C(Background_light)'
formula_intersection = 'Mean_between_sensitivity_and_specificity ~ C(Smoothing_encoded):C(Background_blood):C(Background_light) '

calculate_anova(formula_whole, 'ANOVA')
#calculate_ols(formula_whole, 'OLS', sort=1)
#calculate_ols(formula_whole, 'OLS', sort=2)

#calculate_ols(formula_smoothing_separate, 'OLS_Smoothing')
#calculate_ols(formula_blood_separate, 'OLS_blood')
#calculate_ols(formula_light_separate, 'OLS_light')

#calculate_ols(formula_intersection, 'OLS_intersection', sort=1)
#calculate_ols(formula_intersection, 'OLS_intersection', sort=2)
```
