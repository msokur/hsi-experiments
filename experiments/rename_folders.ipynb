```python
import os

logs_folder = 'D:\\mi186veva-logs'
results_folder = 'D:\\mi186veva-results'

dim = 1

# Function to rename folders
def rename_folders(base_path):
    for folder_name in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder_name)
        if os.path.isdir(folder_path):
            new_folder_name = folder_name.replace('S1_', '')
            new_folder_name = new_folder_name.replace('WT', f'WT_S{dim}')
            new_folder_name = new_folder_name.replace('WF', f'WF_S{dim}')
            new_folder_name = new_folder_name.replace('SS', f'Ss')
            
            new_folder_path = os.path.join(base_path, new_folder_name)

            #os.rename(folder_path, new_folder_path)
            print(f"Renamed: {folder_name} -> {new_folder_name}")

#name = f'MainExperiment_3_smoothing_{dim}d'
name = f'MainExperiment_3_smoothing_savgol'
rename_folders(os.path.join(logs_folder, name))
rename_folders(os.path.join(results_folder, name))
```
